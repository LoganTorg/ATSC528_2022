```python
### Import Required Modules (shouldn't need to change) ###
import numpy as np                 #numpy for math
import matplotlib.pyplot as plt    #matplotlib for plotting
import cartopy.crs as ccrs         #cartopy for plotting on map
import cartopy.feature as cfeature #cartopy basic shapefiles
```

```python
# define variables 

a = 0.25
b = 4
F = 8
G = 1
x0 = np.array([2.5,2.49])
y0 = np.array([1.,1.])
z0 = np.array([0.,0.])
t = 4/(120)  #equeations normalized to 5 days so time step also adjusted

end_time = 120/5 #normalized time step to 5 days, time adjusted
time = np.arange(0,end_time,t) #start,end,interval
```

```python
#creating empty arrays to fill later
x = np.zeros((len(time), len(x0)))
y = np.zeros((len(time), len(y0)))
z = np.zeros((len(time), len(z0)))

#filling first time step with initial conditions
x[0,0] = x0[0]
x[0,1] = x0[1]

y[0,0] = y0[0]
y[0,1] = y0[1]

z[0,0] = z0[0]
z[0,1] = z0[1]

for j in range(len(x0)):
    for i in range(len(time)-1):
        x[(i+1),j] = x[i,j] + (t * (-((y[i,j])**2) - ((z[i,j])**2) - (a * x[i,j]) + (a*F)))
        y[(i+1),j] = y[i,j] + (t * ((x[i,j]*y[i,j]) - (b*x[i,j]*z[i,j]) - y[i,j] + G))
        z[(i+1),j] = z[i,j] + (t * ((b*x[i,j]*y[i,j]) + (x[i,j]*z[i,j]) - z[i,j]))
        
             
#print(z)
```

```python
#calculating the differences between the simulations 
dif_x = x[:,1] - x[:,0]
dif_y = y[:,1] - y[:,0]
dif_z = z[:,1] - z[:,0]
```

```python
#plotting simulation with x value of 2.5

fig = plt.figure(figsize=(15,5),dpi=200)
ax1 = fig.add_subplot(131)
ax1.plot(time*5, x[:,0])
plt.ylabel('strenght of circumpolar westerly current')
plt.xlabel('Time (days)')
plt.grid()

ax1 = fig.add_subplot(132)
ax1.plot(time*5, y[:,0])
plt.ylabel('cosine phases')
plt.xlabel('Time (days)')
plt.grid()

ax1 = fig.add_subplot(133)
ax1.plot(time*5, z[:,0])
plt.ylabel('sine phases')
plt.xlabel('Time (days)')
plt.grid()

plt.suptitle('Simulation 1')
plt.tight_layout()
#plt.savefig("C:\\Users\\torge\\Documents\\GitHub\\ATSC528_2022\\04-Chaos\\sim_1.png")
```

```python
#plotting our error value 2.49 instead of 2.5

fig = plt.figure(figsize=(15,5),dpi=200)
ax1 = fig.add_subplot(131)
ax1.plot(time*5, x[:,1])
plt.ylabel('strenght of circumpolar westerly current')
plt.xlabel('Time (days)')
plt.grid()

ax1 = fig.add_subplot(132)
ax1.plot(time*5, y[:,1])
plt.ylabel('cosine phases')
plt.xlabel('Time (days)')
plt.grid()

ax1 = fig.add_subplot(133)
ax1.plot(time*5, z[:,1])
plt.ylabel('sine phases')
plt.xlabel('Time (days)')
plt.grid()

plt.suptitle('Simulation 2')
plt.tight_layout()
#plt.savefig("C:\\Users\\torge\\Documents\\GitHub\\ATSC528_2022\\04-Chaos\\sim_2.png")
```

```python
#making difference plots- difference between simulation 1 and 2 

fig = plt.figure(figsize=(15,5),dpi=200)
ax1 = fig.add_subplot(131)
ax1.plot(time*5, dif_x)
plt.ylabel('strenght of circumpolar westerly current')
plt.xlabel('Time (days)')
plt.grid()

ax1 = fig.add_subplot(132)
ax1.plot(time*5, dif_y)
plt.ylabel('cosine phases')
plt.xlabel('Time (days)')
plt.grid()

ax1 = fig.add_subplot(133)
ax1.plot(time*5, dif_z)
plt.ylabel('sine phases')
plt.xlabel('Time (days)')
plt.grid()

plt.suptitle('Difference Plot')
plt.tight_layout()
#plt.savefig("C:\\Users\\torge\\Documents\\GitHub\\ATSC528_2022\\04-Chaos\\dif_plot.png")
```

```python
#Comparison Plots

fig = plt.figure(figsize=(15,5),dpi=200)
ax1 = fig.add_subplot(131)
ax1.plot(time*5, x[:,0])        #have to do this to get the axis to say 125 days
ax1.plot(time*5, x[:,1])
ax1.legend(['Simulation 1', 'Simulation 2'])
plt.ylabel('strenght of circumpolar westerly current')
plt.xlabel('Time (days)')
plt.grid()

ax1 = fig.add_subplot(132)
ax1.plot(time*5, y[:,0])
ax1.plot(time*5, y[:,1])
ax1.legend(['Simulation 1', 'Simulation 2'])
plt.ylabel('cosine phases')
plt.xlabel('Time (days)')
plt.grid()

ax1 = fig.add_subplot(133)
ax1.plot(time*5, z[:,0])
ax1.plot(time*5, z[:,1])
ax1.legend(['Simulation 1', 'Simulation 2'])
plt.ylabel('sine phases')
plt.xlabel('Time (days)')
plt.grid()

plt.suptitle('Comparison Plots')
plt.tight_layout()
#plt.savefig("C:\\Users\\torge\\Documents\\GitHub\\ATSC528_2022\\04-Chaos\\Comparison_plots.png")
```
