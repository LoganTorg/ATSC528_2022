```python
### Import Required Modules (shouldn't need to change) ###
import numpy as np                 #numpy for math
import matplotlib.pyplot as plt    #matplotlib for plotting
import cartopy.crs as ccrs         #cartopy for plotting on map
import cartopy.feature as cfeature #cartopy basic shapefiles
import pandas as pd                #pandas to open file
```

```python
### Create function for Cressman Analysis ###
def cressman(dik, R):
    if dik <= R:
        w = (R**2-dik**2)/(R**2+dik**2)   #equation given
        
    elif dik >R:
        w = 0
            
    return w
```

```python
### Create function for bilinear interpolation ###
def bilinear(x,y,ROI):
    rel_ind_x = x-np.floor(x)
    rel_ind_y = y-np.floor(y)
    
    z1 = analysis_vals[int(np.floor(y)),int(np.floor(x)),ROI]
    z2 = analysis_vals[int(np.ceil(y)),int(np.floor(x)),ROI]
    z3 = analysis_vals[int(np.ceil(y)),int(np.ceil(x)),ROI]
    z4 = analysis_vals[int(np.floor(y)),int(np.ceil(x)),ROI]
   
    p = (((rel_ind_x-1)*(rel_ind_y-1)*z1) - ((rel_ind_x-0)*(rel_ind_y-1)*z2) + ((rel_ind_x-0)*(rel_ind_y-0)*z3) - ((rel_ind_x-1)*(rel_ind_y-0)*z4))
    
    return p
    
```

```python
### Read in observations ###

file = open("C:\\Users\\torge\\Documents\\GitHub\\ATSC528_2022\\RAOBs_201903131200.txt", "r")
data = pd.read_csv(file, sep=",")
data[data.columns[1:5]] = data[data.columns[1:5]].astype(float) #Converting the data in the text file into numbers(float). Locations stay as letters.
#print(contents)

### Define Variables ###
xo = 18.9
yo = -6.3
delta_x = 1.27
delta_y = 1.27
phi_o = 60
lambda_o = -115
rho = 637100000        #Converted to centimeters 
map_scale = 1/15000000
fo = data['Height'].values #500mb height values
fo = np.array(fo)          #must do this because we are using pandas
```

```python
### Set up analysis map with a 22x28 rectangular grid of points ###

l = xo+np.arange(22) * delta_x      #making an x,y grid
k = yo+np.arange(28) * delta_y
x_grid, y_grid = np.meshgrid(l, k)

grid_lon = np.arctan(y_grid/x_grid)*(180/np.pi)+lambda_o      #Converting our grid points from x,y to lat,lon so we can overlay it onto a map projection
grid_lat=(180/np.pi)*((np.pi/2)-(2*np.arctan(np.sqrt((x_grid/map_scale)**2+(y_grid/map_scale)**2)/(rho*(1+np.sin(phi_o*np.pi/180.)))))) 

```

```python
### convert obs lat/long to x,y ###

#Converting our observations from lat,lon to x,y grid because Jared said so
sigma = ((1.0+(np.sin(phi_o*np.pi / 180.))) / (1.0+(np.sin(data['Lat'] * np.pi / 180.))))  #Have to convert to radians for calculation by using pi/180
x = sigma * rho * np.cos(data['Lat'] * np.pi / 180.) * np.cos((data['Lon']-lambda_o) * np.pi / 180.) * map_scale  #Have to multilpy by map scale otherwise values would be in terms of the entire earth
y = sigma * rho * np.cos(data['Lat'] * np.pi / 180.) * np.sin((data['Lon']-lambda_o) * np.pi / 180.) * map_scale

x = np.array(x)     #making our x,y obs into an array to avoid errors
y = np.array(y)

obs_lon = np.arctan(y/x)*(180/np.pi)+lambda_o   #Converting our observations back into lat,lon so we can plot and make sure our grid and overlayed observations are correct
obs_lat = (180/np.pi)*((np.pi/2)-(2*np.arctan(np.sqrt((x/map_scale)**2+(y/map_scale)**2)/(rho*(1+np.sin(phi_o*np.pi/180.))))))

proj = ccrs.Stereographic(central_longitude=-115,central_latitude=90,true_scale_latitude=60)
fig = plt.figure(figsize=(8,8),dpi=200)
ax1 = fig.add_subplot(111,projection=proj)
ax1.add_feature(cfeature.STATES)
ax1.add_feature(cfeature.COASTLINE)
ax1.scatter(grid_lon.ravel(), grid_lat.ravel(),transform = ccrs.PlateCarree())
ax1.scatter(obs_lon, obs_lat, transform = ccrs.PlateCarree())

#plt.savefig("C:\\Users\\torge\\Documents\\GitHub\\ATSC528_2022\\02-Succesive_Corrections\\Analysis and Obs Field.png")   #Have to save this figure


#plt.show()
```

```python
### Perform 500mb geopotential height analyses using a Cressman weighting Function###
#Use radii of influence 4, 2.5, 1.5 *dmin


dist = np.empty(len(x))     #x is our x_ob                #creating a list full of the minimum distances from each point
for i in range(len(x)):
    R = (((x[i]-x)**2) + ((y[i]-y)**2))**(1/2)            #using pathagorean theorem to find the closest observation
    R = np.delete(R, i)                                      #the closest observation will always be itself
    dist[i] = np.min(R)                            #this will store all the minimums in one list
    
#we need dmin which is average of the minimum distances
dmin = np.mean(dist)

#define our ROIs
ROI1 = 4*dmin
ROI2 = 2.5*dmin
ROI3 = 1.5*dmin
#print(dmin)

#need an ROI matrix to call later
ROI_mat = np.array([ROI1,ROI2,ROI3]) 
```

```python
### First analysis, no successive corrections ###
#in our first pass we dont have any background

y_n = 22                          #this is the size of our grid
x_n = 28
analysis_vals = np.zeros((x_n,y_n,len(ROI_mat)))  

#for the first pass we only use the first ROI
for m in range(len(x_grid)):
    for n in range(len(x_grid[0])):
        rel_x = x_grid[m,n] - x      #subtracting our obs from our analysis to get a relative distance for our map
        rel_y = y_grid[m,n] - y
        radius = (rel_x**2 + rel_y**2)**(1/2) #radius of influence for our ith analysis point
        index_roi = np.where(radius <=ROI_mat[0])[0] #only obs within roi will affect analysis value
            
        #now we need to calculate our weights
        weights = np.zeros(len(x))       #every observation has its own weight
        for i in index_roi:
            w = cressman(radius[i], ROI_mat[0])    #calculating cressman using previous function
            weights[i] = w                         #storing our weights in a list
            
            #now we need the analysis values at the grid points
        analysis_vals[m,n,0] = np.sum(weights*fo)/np.sum(weights)   #this is our first pass
```

```python
### Next Analyses ###
#Second Pass#

ind_x = (x-xo)/delta_x
ind_y = (y-yo)/delta_y

p = np.zeros(len(x))
for i in range(len(ind_x)):   #can only do bilinear for obs that are in the analysis grid
    if (ind_x[i] >= 0 and ind_x[i] <= 21) and (ind_y[i] >= 0 and ind_y[i] < 27):
        P = bilinear(ind_x[i], ind_y[i], 0) #we use 0 because that is the only part we have filled so far
        
        #if the obs are outside of our grid we need to do a "backward cressman" (whatever that is)
    else:
        rel_xa = x[i] - x_grid ##subtract grid from obs
        rel_ya = y[i] - y_grid ##relative distance on map
        radius_a = (rel_xa**2 + rel_ya**2)**(1/2) #ROI for obs point
        index_roi_a = np.where(radius_a <= ROI_mat[0])
        
        #need a weight function again, but now for giving weights to analysis to get the background values 
        weights_a = np.zeros(x_grid.shape)
        for k in index_roi_a[0]:
            for j in index_roi_a[1]:
                w_a = cressman(radius_a[k,j], ROI_mat[0])
                weights_a[k,j] = w_a
        P = np.sum(weights_a*analysis_vals[:,:,0]) / np.sum(weights_a)
        
    p[i] = P #analysis values at obs locations
    
# need new analysis values at grid points
for m in range(len(x_grid)):
    for n in range(len(x_grid[0])):
        rel_x = x_grid[m,n] - x   #subtracting obs from analysis
        rel_y = y_grid[m,n] - y
        radius = (rel_x**2 + rel_y**2)**(1/2) #ROI for ith analysis point
        index_roi = np.where(radius <= ROI_mat[1])[0]
        
        #get weights for each ob to influence grid points
        weights = np.zeros(len(x))
        for i in index_roi:
            w = cressman(radius[i], ROI_mat[1])
            weights[i] = w
            
        #these are the analysis values at the grid points (f_A^2 (r_i))
        analysis_vals[m,n,1] = analysis_vals[m,n,0] + (np.nansum(weights * (fo-p)) / np.nansum(weights))   

```

```python
#Third Pass

pp = np.zeros(len(x))
for i in range(len(ind_x)):   #can only do bilinear for obs that are in the analysis grid
    if (ind_x[i] >= 0 and ind_x[i] <= 21) and (ind_y[i] >= 0 and ind_y[i] < 27):
        P = bilinear(ind_x[i], ind_y[i], 1) #we use 0 because that is the only part we have filled so far
        
        #if the obs are outside of our grid we need to do a "backward cressman" (whatever that is)
    else:
        rel_xa = x[i] - x_grid ##subtract grid from obs
        rel_ya = y[i] - y_grid ##relative distance on map
        radius_a = (rel_xa**2 + rel_ya**2)**(1/2) #ROI for obs point
        index_roi_a = np.where(radius_a <= ROI_mat[1])
        
        #need a weight function again, but now for giving weights to analysis to get the background values 
        weights_a = np.zeros(x_grid.shape)
        for k in index_roi_a[0]:
            for j in index_roi_a[1]:
                w_a = cressman(radius_a[k,j], ROI_mat[1])
                weights_a[k,j] = w_a
        P = np.sum(weights_a*analysis_vals[:,:,1]) / np.sum(weights_a)
        
    pp[i] = P #analysis values at obs locations
    
# need new analysis values at grid points
for m in range(len(x_grid)):
    for n in range(len(x_grid[0])):
        rel_x = x_grid[m,n] - x   #subtracting obs from analysis
        rel_y = y_grid[m,n] - y
        radius = (rel_x**2 + rel_y**2)**(1/2) #ROI for ith analysis point
        index_roi = np.where(radius <= ROI_mat[2])[0]
        
        #get weights for each ob to influence grid points
        weights = np.zeros(len(x))
        for i in index_roi:
            w = cressman(radius[i], ROI_mat[2])
            weights[i] = w
            
        #these are the analysis values at the grid points (f_A^2 (r_i))
        analysis_vals[m,n,2] = analysis_vals[m,n,1] + (np.nansum(weights * (fo-pp)) / np.nansum(weights))   
#print(analysis_vals)
```

```python
### Calculate Analysis Differences

diff = np.zeros((x_n,y_n,len(ROI_mat)))  #calculating the difference between the analysis values from the different passes
diff[:,:,0] = analysis_vals[:,:,1] - analysis_vals[:,:,0]   #differene between first and second pass
diff[:,:,1] = analysis_vals[:,:,2] - analysis_vals[:,:,0]   #difference between third and first pass
diff[:,:,2] = analysis_vals[:,:,2] - analysis_vals[:,:,1]   #difference between third and second pass
```

```python
#Calculating root mean squares

RMS = np.zeros((len(ROI_mat)))

#first pass
first_pass_diff = (fo-p)**2      #background is the analysis values from the second pass
first_pass_diff = first_pass_diff[~np.isnan(first_pass_diff)] # ~ actually deletes the values instead of keeping them
RMS[0] = np.sqrt((np.sum(first_pass_diff))/len(first_pass_diff))

#second pass
sec_pass_diff = (fo-pp)**2     #background is the analysis values from the third pass
sec_pass_diff = sec_pass_diff[~np.isnan(sec_pass_diff)]
RMS[1] = np.sqrt((np.sum(sec_pass_diff))/len(sec_pass_diff))

#third pass
#need to calculate f_A^3(r_k) from our background values
ppp = np.zeros(len(x))
for i in range(len(ind_x)):   #can only do bilinear for obs that are in the analysis grid
    if (ind_x[i] >= 0 and ind_x[i] <= 21) and (ind_y[i] >= 0 and ind_y[i] < 27):
        P = bilinear(ind_x[i], ind_y[i], 2) #we use 0 because that is the only part we have filled so far
        
        #if the obs are outside of our grid we need to do a "backward cressman" (whatever that is)
    else:
        rel_xa = x[i] - x_grid ##subtract grid from obs
        rel_ya = y[i] - y_grid ##relative distance on map
        radius_a = (rel_xa**2 + rel_ya**2)**(1/2) #ROI for obs point
        index_roi_a = np.where(radius_a <= ROI_mat[2])
        
        #need a weight function again, but now for giving weights to analysis to get the background values 
        weights_a = np.zeros(x_grid.shape)
        for k in index_roi_a[0]:
            for j in index_roi_a[1]:
                w_a = cressman(radius_a[k,j], ROI_mat[2])
                weights_a[k,j] = w_a
        P = np.sum(weights_a*analysis_vals[:,:,2]) / np.sum(weights_a)
        
    ppp[i] = P #analysis values at obs locations
    
#calculate RMS
third_pass_diff = (fo-ppp)**2           #background is analysis values (f_A^3 from above
third_pass_diff = third_pass_diff[~np.isnan(third_pass_diff)]
RMS[2] = np.sqrt((np.sum(third_pass_diff))/len(third_pass_diff))
#print(RMS)
 
```
