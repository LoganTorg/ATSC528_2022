```python
"""
Created By    : Jared W. Marquis
Creation Date : 01 August 2022
Course        : ATSC 528 - Atmospheric Data Analysis
Assignment    : #03 - Statistical Objective Analysis

Purpose:
In this assignment we want to take upper air observations and perform perform an analysis using SOA. We could do this using successive corrections
but that might require infinite passes to reach the optimal nalysis while SOA will do this with one pass. In SOA the analysis value is equal
to the background value plus a weighted average of observation increments. Using the Barnes analysis we needed to find the background at the obs
and grid points. From here we could use our known information to plug into the analysis equation. This allowed me to plot the data on a 500mb
polar stereographic map and analyze the data. I then calculated the analysis increment (the increment between the analysis using 2/4/10 closest 
analysis and background at that grid point). This shows how much these analyses differ from the background field. Calculating the analysis differences
gives us a good look at how good our analysis is, this is only possible because we assume our observations are perfect. THe last step was to calculate
the RMS values to see how much error our analysis contains. 
"""
__author__    = "Logan Torgerson"
__contact__   = "logan.torgerson@und.edu"
```

```python
### Import Required Modules (shouldn't need to change) ###
import numpy as np                 #numpy for math
import matplotlib.pyplot as plt    #matplotlib for plotting
import cartopy.crs as ccrs         #cartopy for plotting on map
import cartopy.feature as cfeature #cartopy basic shapefiles
import pandas as pd                #pandas to open file
```

```python
### Create function for bilinear interpolation ###
def bilinear(x,y):
    rel_ind_x = x-np.floor(x)
    rel_ind_y = y-np.floor(y)
    
    z1 = background_grid[int(np.floor(y)),int(np.floor(x))]
    z2 = background_grid[int(np.ceil(y)),int(np.floor(x))]
    z3 = background_grid[int(np.ceil(y)),int(np.ceil(x))]
    z4 = background_grid[int(np.floor(y)),int(np.ceil(x))]
   
    p = (((rel_ind_x-1)*(rel_ind_y-1)*z1) - ((rel_ind_x-0)*(rel_ind_y-1)*z2) + ((rel_ind_x-0)*(rel_ind_y-0)*z3) - ((rel_ind_x-1)*(rel_ind_y-0)*z4))
    
    return p
```

```python
### Bilinear interpolation for analysis difference/forward interpolation ###

def bilinear2(x,y,N):
        rel_ind_x = x-np.floor(x)
        rel_ind_y = y-np.floor(y)

        z1 = analysis_mat[int(np.floor(y)),int(np.floor(x)),N]
        z2 = analysis_mat[int(np.ceil(y)),int(np.floor(x)),N]
        z3 = analysis_mat[int(np.ceil(y)),int(np.ceil(x)),N]
        z4 = analysis_mat[int(np.floor(y)),int(np.ceil(x)),N]

        p = (((rel_ind_x-1)*(rel_ind_y-1)*z1) - ((rel_ind_x-0)*(rel_ind_y-1)*z2) + ((rel_ind_x-0)*(rel_ind_y-0)*z3) - ((rel_ind_x-1)*(rel_ind_y-0)*z4))

        return p
```

```python
### Creating a function for the Barnes Function ###
#copied from successive corrections

def barnes(dik):     #barnes only has one variable since kd and r are set
    w = np.exp((-(dik**2))/(kd))   #equation given
    return w
```

```python
# The height data and grid we are using is the same from previous assignments, this part copied and pasted

### Define Variables ###
xo = 18.9
yo = -6.3
delta_x = 1.27
delta_y = 1.27
phi_o = 60
lambda_o = -115
rho = 637100000        #Converted to centimeters 
circum = 6371          #circumference of earth in meters
map_scale = 1/15000000
re = 12.7775892        #units cm
kd = 10.8844524        #smoothing parameter for barnes units cm^2

N_arr = np.array([2,4,10])      #list containing the N closest stations

### Read in Observations ###
file = open("C:\\Users\\torge\\Documents\\GitHub\\ATSC528_2022\\RAOBs_201903131200.txt", "r")
data = pd.read_csv(file, sep=",")
data[data.columns[1:5]] = data[data.columns[1:5]].astype(float) #Converting the data in the text file into numbers(float). Locations stay as letters.
#print(contents)

fo = data['Height'].values #500mb height values
fo = np.array(fo)          #must do this because we are using pandas
```

```python
### Set up analysis map with a 22x28 rectangular grid of points ###
#Copied from previous homeworks
l = xo+np.arange(22) * delta_x      #making an x,y grid
k = yo+np.arange(28) * delta_y
x_grid, y_grid = np.meshgrid(l, k)

grid_lon = np.arctan(y_grid/x_grid)*(180/np.pi)+lambda_o      #Converting our grid points from x,y to lat,lon so we can overlay it onto a map projection
grid_lat=(180/np.pi)*((np.pi/2)-(2*np.arctan(np.sqrt((x_grid/map_scale)**2+(y_grid/map_scale)**2)/(rho*(1+np.sin(phi_o*np.pi/180.)))))) 
rad_lat = grid_lat *np.pi/180   #need to make the grid in radians as well to calculate s
rad_lon = grid_lon *np.pi/180
```

```python
### convert obs lat/long to x,y (may want to plot on your analysis grid to verify)###
#copied from previous homeworks

#Converting our observations from lat,lon to x,y grid because Jared said so
sigma = ((1.0+(np.sin(phi_o*np.pi / 180.))) / (1.0+(np.sin(data['Lat'] * np.pi / 180.))))  #Have to convert to radians for calculation by using pi/180
x = sigma * rho * np.cos(data['Lat'] * np.pi / 180.) * np.cos((data['Lon']-lambda_o) * np.pi / 180.) * map_scale  #Have to multilpy by map scale otherwise values would be in terms of the entire earth
y = sigma * rho * np.cos(data['Lat'] * np.pi / 180.) * np.sin((data['Lon']-lambda_o) * np.pi / 180.) * map_scale

x = np.array(x)     #making our x,y obs into an array to avoid errors
y = np.array(y)

obs_lon = np.arctan(y/x)*(180/np.pi)+lambda_o   #Converting our observations back into lat,lon so we can plot and make sure our grid and overlayed observations are correct
obs_lat = (180/np.pi)*((np.pi/2)-(2*np.arctan(np.sqrt((x/map_scale)**2+(y/map_scale)**2)/(rho*(1+np.sin(phi_o*np.pi/180.))))))

proj = ccrs.Stereographic(central_longitude=-115,central_latitude=90,true_scale_latitude=60)
fig = plt.figure(figsize=(8,8),dpi=200)
ax1 = fig.add_subplot(111,projection=proj)
ax1.add_feature(cfeature.STATES)
ax1.add_feature(cfeature.COASTLINE)
ax1.scatter(grid_lon.ravel(), grid_lat.ravel(),transform = ccrs.PlateCarree())
ax1.scatter(obs_lon, obs_lat, transform = ccrs.PlateCarree())

#plt.savefig("C:\\Users\\torge\\Documents\\GitHub\\ATSC528_2022\\01-Function_Fitting\\Analysis and Obs Field.png")   #Have to save this figure


plt.show()
```

```python
###Using barnes weighting function to get background values at grid points ###


y_n = 22                          #this is the size of our grid
x_n = 28
background_grid = np.zeros((x_n,y_n)) 


for m in range(len(x_grid)):
    for n in range(len(x_grid[0])):
        rel_x = x_grid[m,n] - x      #subtracting our obs from our analysis to get a relative distance for our map
        rel_y = y_grid[m,n] - y
        radius = (rel_x**2 + rel_y**2)**(1/2) #radius of influence for our ith analysis point
        index_roi = np.where(radius <= re)[0] #only obs within roi will affect analysis value
            
        #now we need to calculate our weights
        weights = np.zeros(len(x))       #every observation has its own weight
        for i in index_roi:
            w = barnes(radius[i])    #calculating cressman using previous function
            weights[i] = w                         #storing our weights in a list
        W = weights/np.nansum(weights)
            
            #now we need the background values at grid points
        background_grid[m,n] = (np.sum(W*fo)) #this is our background
```

```python
#now need to find the background values for the observation locations

ind_x = (x-xo)/delta_x
ind_y = (y-yo)/delta_y

p = np.zeros((len(x)))
for i in range(len(ind_x)):   #can only do bilinear for obs that are in the analysis grid
    if (ind_x[i] >= 0 and ind_x[i] <= 21) and (ind_y[i] >= 0 and ind_y[i] < 27):
        P = bilinear(ind_x[i], ind_y[i])
        
        #if the obs are outside of our grid we need to do a "backward cressman" (whatever that is)
    else:
        rel_xa = x[i] - x_grid ##subtract grid from obs
        rel_ya = y[i] - y_grid ##relative distance on map
        radius_a = (rel_xa**2 + rel_ya**2)**(1/2) #ROI for obs point
        index_roi_a = np.where(radius_a <= re)
        
        #need a weight function again, but now for giving weights to analysis to get the background values 
        weights_a = np.zeros(x_grid.shape)
        for k in index_roi_a[0]:
            for j in index_roi_a[1]:
                w_a = barnes(radius_a[k,j])
                weights_a[k,j] = w_a
        W = weights_a/np.nansum(weights_a)
        P = (np.sum(W*background_grid))  #calculating the background at the obs lo*analysis_mat[:,:,0]
    p[i] = P #putting the values into a list
#print(p)
    
```

```python
#need to calculate the analysis

analysis_mat = np.zeros((x_n,y_n,len(N_arr)))    #create empty list to put values into 
for i in range(len(N_arr)):
    analysis_arr = N_arr[i]               
    for m in range(len(x_grid)):       #looping through our grid
        for n in range(len(x_grid[0])):
            rel_x = x_grid[m,n] - x      #subtracting our obs from our analysis to get a relative distance for our map
            rel_y = y_grid[m,n] - y
            radius = (rel_x**2 + rel_y**2)**(1/2) #even though we are not using ROIs we still need to know the distances
            station_sort = np.argsort(radius)    #to figure out the closest stations (N values) 
            station = station_sort[:analysis_arr] #we will use either 2,4,10 closest stations
            
            new_lat = obs_lat[station] *np.pi/180   #we need to change our values into radians to calculate s
            new_lon = obs_lon[station] *np.pi/180
            new_heights = fo[station]
            new_background = p[station] 
            
            #distance from ob point to ob point and from ob point to grid point
            dist_ob_ob =  np.zeros((len(station), len(station)))       #create an empty grid to store dist values
            dist_ob_grid =  np.zeros((len(station)))
            for j in range(len(new_lat)):
                s = 2 * circum * np.arcsin(np.sqrt(((np.sin((new_lat-new_lat[j])/2))**2)+(np.cos(new_lat[j])*np.cos(new_lat)*((np.sin((new_lon-new_lon[j])/2))**2)))) #get the distance between obs on earth / s values for the analysis equation 
                dist_ob_ob[:,j] = s        #saving calculated values
                
                s = 2 * circum * np.arcsin(np.sqrt(((np.sin((new_lat[j]-rad_lat[m,n])/2))**2)+(np.cos(rad_lat[m,n])*np.cos(new_lat[j])*((np.sin((new_lon[j]-rad_lon[m,n])/2))**2)))) #get the distance between obs and grid on earth  
                dist_ob_grid[j] = s   
                
                #these distances need to be changed to 1000s of km for the correlation equation
                ob_ob = dist_ob_ob / 1000
                ob_grid = dist_ob_grid / 1000
                
                #now we can actually calculate the correlation equation
                cor_ob_ob = np.zeros((len(station), len(station)))
                cor_ob_grid = np.zeros((len(station), len(station)))
                
                #calcualting the correlation between ob to ob
                cor1 = 0.95 * np.exp ((-1.24)*(ob_ob**2))
                cor1[cor1 >= 0.9499] = 1   #the max correlation allowed by the equation is .95 when the correlation between an ob and itself should be 1
                cor1 = np.matrix(cor1)            #must convert data into matrix
                
                #calcuating the correlation between ob to grid
                cor2 = 0.95 * np.exp ((-1.24)*(ob_grid**2))
                cor2[cor2 >= 0.9499] = 1
                cor2 = np.matrix(cor2).T     #taking the transpose of matrix
                
                weights = np.linalg.inv(cor1) * cor2    #inverting the matrix
                weights = np.array(weights).flatten()
                
                analysis_mat[m,n,i] = background_grid[m,n] + np.nansum((weights * (new_heights-new_background)))  #calculating the analysis equation
                                                     
```

```python
#need to get the analysis values at obs locations

#creating a empty matrix to be filled later
analysis_vals_obs = np.zeros((len(x), len(N_arr)))  #analysis values at observation locations

        #need to do bilinear if obs are in our grid and forward interpolation if outside our grid
for m in range(len(N_arr)):
    for i in range(len(ind_x)):   #can only do bilinear for obs that are in the analysis grid
        if (ind_x[i] >= 0 and ind_x[i] <= 21) and (ind_y[i] >= 0 and ind_y[i] < 27):
            P = bilinear2(ind_x[i], ind_y[i], m) #we use 0 because that is the only part we have filled so far

            #if the obs are outside of our grid we need to do a "reverse barnes" 
        else:
            rel_xa = x[i] - x_grid ##subtract grid from obs
            rel_ya = y[i] - y_grid ##relative distance on map
            radius_a = (rel_xa**2 + rel_ya**2)**(1/2) #ROI for obs point
            index_roi_a = np.where(radius_a <= re)

            #need a weight function again, but now for giving weights to analysis to get the background values 
            weights_a = np.zeros(x_grid.shape)
            for k in index_roi_a[0]:
                for j in index_roi_a[1]:
                    w_a = barnes(radius_a[k,j])
                    weights_a[k,j] = w_a
            W = weights_a/np.nansum(weights_a)
            P = np.sum(W * analysis_mat[:,:,m])  #calculating the background at the obs locations
            #print(P)  
        analysis_vals_obs[i,m] = P #putting the values into a list
    #print(P)
```

```python
### Calculate Analysis Differences

diff = np.zeros((x_n,y_n,len(N_arr)))  #calculating the difference between the analysis values from the different passes
diff[:,:,0] = analysis_mat[:,:,1] - analysis_mat[:,:,0]   #differene between 
diff[:,:,1] = analysis_mat[:,:,2] - analysis_mat[:,:,0]   #difference between 
diff[:,:,2] = analysis_mat[:,:,2] - analysis_mat[:,:,1]   #difference between 
```

```python
#Calculating root mean squares differences between analysis and obs at obs locations

RMS = np.zeros((len(N_arr)))

#N=2
N2_dif = (fo-analysis_vals_obs[:,0])**2      
N2_dif = N2_dif[~np.isnan(N2_dif)] # deleting nan values
RMS[0] = np.sqrt((np.sum(N2_dif))/len(N2_dif))

#N=4
N4_dif = (fo-analysis_vals_obs[:,1])**2      
N4_dif = N4_dif[~np.isnan(N4_dif)] # deleting nan values
RMS[1] = np.sqrt((np.sum(N4_dif))/len(N4_dif))

#N=10
N10_dif = (fo-analysis_vals_obs[:,2])**2      
N10_dif = N10_dif[~np.isnan(N10_dif)] # deleting nan values
RMS[2] = np.sqrt((np.sum(N10_dif))/len(N10_dif))
    
print(RMS)
```
